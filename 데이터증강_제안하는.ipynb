```python
# 소수구간에 해당하는 데이터를 3배 또는 2배로 증강
def aug(x):
  result = pd.DataFrame([],columns=x.columns)
  plus = pd.DataFrame([],columns=x.columns)

  for idx, row in x.iterrows():
    # 3배로 증강
    if 1000 <= row['Roller1_speed'] <=1200 or 1501 <= row['Roller1_speed'] or row['Roller2_temp'] != 100:
      plus = plus.append(row, ignore_index=True)
      plus = plus.append(row, ignore_index=True)
      print(row)
    # 2배로 증강
    elif row['Roller2_speed'] != 4105 or row['FR_speed'] != 4000:
      plus = plus.append(row, ignore_index=True)
      print(row)
  
  print(plus)
  result = x.append(plus, ignore_index=True)
  return result
```
